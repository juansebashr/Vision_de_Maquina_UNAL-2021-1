### 2.   Implementar, como una función para OpenCV, las conversiones entre los espacios de color siguientes:
* RGB a LUX y su inversa.
* RGB a OHTA y su inversa.

###evaluarlas en su proyecto o en aplicaciones a detección de rostros y de segmentación de productos, respectivamente.

---

---

```
def RGB2LUX(img):
  N,M = img.shape[:2]
  R = img[:,:,0]
  G = img[:,:,1]
  B = img[:,:,2]
  L = np.zeros((N,M))
  U = np.zeros((N,M))
  X = np.zeros((N,M))
  for i in range(N):
    for j in range(M):
      L[i,j] = ((np.sign(R[i,j]-1)*np.power(np.abs(R[i,j]-1),0.3))*\
                (np.sign(G[i,j]-1)*np.power(np.abs(G[i,j]-1),0.6))*\
                (np.sign(B[i,j]-1)*np.power(np.abs(B[i,j]-1),0.1))) - 1
      if R[i,j] < L[i,j]:
        U[i,j] = (256/2)*((R[i,j]+1)/(L[i,j]+1))
      else:
        U[i,j] = 256-(256/2)*((L[i,j]+1)/(R[i,j]+1))
      if B[i,j] < L[i,j]:
        X[i,j] = (256/2)*((B[i,j]+1)/(L[i,j]+1))
      else:
        X[i,j] = 256-(256/2)*((L[i,j]+1)/(B[i,j]+1))
  imgLUX = np.array([L,U,X])
  return imgLUX

def LUX2RGB(img):
  N,M = img.shape[:2]
  L = img[:,:,0]
  U = img[:,:,1]
  X = img[:,:,2]
  R = np.zeros((N,M))
  G = np.zeros((N,M))
  B = np.zeros((N,M))
  imgRGB = np.array([R,G,B])
  return imgRGB

def RGB2OHTA(img):
  N,M = img.shape[:2]
  R = img[:,:,0]
  G = img[:,:,1]
  B = img[:,:,2]
  I1 = np.zeros((N,M))
  I2 = np.zeros((N,M))
  I3 = np.zeros((N,M))
  T = np.array([[1/3,1/3,1/3],
                [1/2,0,-1/2],
                [-1/4,1/2,-1/4]])
  for i in range(N):
    for j in range(M):
      [I1[i,j],I2[i,j],I3[i,j]] = np.matmul(T,[R[i,j],G[i,j],B[i,j]])
  imgOHTA = np.array([I1,I2,I3])
  return imgOHTA

def OHTA2RGB(img):
  N,M = img.shape[1:]
  I1 = img[0,:,:]
  I2 = img[1,:,:]
  I3 = img[2,:,:]
  R = np.zeros((N,M))
  G = np.zeros((N,M))
  B = np.zeros((N,M))
  T = np.array([[1,1,-2/3],
                [1,0,4/3],
                [1,-1,-2/3]])
  for i in range(N):
    for j in range(M):
      [R[i,j],G[i,j],B[i,j]] = np.matmul(T,[I1[i,j],I2[i,j],I3[i,j]])
  imgRGB = np.array([R,G,B])
  return imgRGB

imgLUX = RGB2LUX(img)
print(imgLUX.shape)
plt.figure(figsize=(20,20))
plt.subplot(131),plt.imshow(imgLUX[0,:,:], cmap=plt.cm.gray), plt.title('Plano L', fontsize = 15)
plt.subplot(132),plt.imshow(imgLUX[1,:,:], cmap=plt.cm.gray), plt.title('Plano U', fontsize = 15)
plt.subplot(133),plt.imshow(imgLUX[2,:,:], cmap=plt.cm.gray), plt.title('Plano X', fontsize = 15)
plt.show()

imgOHTA = RGB2OHTA(img)
print(imgOHTA.shape)
plt.figure(figsize=(20,20))
plt.subplot(131),plt.imshow(imgOHTA[0,:,:], cmap=plt.cm.gray), plt.title('Plano I1', fontsize = 15)
plt.subplot(132),plt.imshow(imgOHTA[1,:,:], cmap=plt.cm.gray), plt.title('Plano I2', fontsize = 15)
plt.subplot(133),plt.imshow(imgOHTA[2,:,:], cmap=plt.cm.gray), plt.title('Plano I3', fontsize = 15)
plt.show()
  
```

Al implementar las conversiones de RGB a LUX y OHTA se observa como los planos LUX son útules para identificación de bordes y resalte, por otro lado los planos OHTA resaltan las regiones de color claro de la imagen.

```
imgRGB = OHTA2RGB(imgOHTA)
plt.figure(figsize=(20,20))
plt.subplot(131),plt.imshow(imgRGB[0,:,:], cmap=plt.cm.gray), plt.title('Plano R', fontsize = 15)
plt.subplot(132),plt.imshow(imgRGB[1,:,:], cmap=plt.cm.gray), plt.title('Plano G', fontsize = 15)
plt.subplot(133),plt.imshow(imgRGB[2,:,:], cmap=plt.cm.gray), plt.title('Plano B', fontsize = 15)
plt.show()

plt.figure(figsize=(20,20))
plt.subplot(131),plt.imshow(img[:,:,0], cmap=plt.cm.gray), plt.title('Plano R original', fontsize = 15)
plt.subplot(132),plt.imshow(img[:,:,1], cmap=plt.cm.gray), plt.title('Plano G original', fontsize = 15)
plt.subplot(133),plt.imshow(img[:,:,2], cmap=plt.cm.gray), plt.title('Plano B original', fontsize = 15)
plt.show()

plt.figure(figsize=(20,20))
plt.subplot(131),plt.imshow(img[:,:,0]-imgRGB[0,:,:], cmap=plt.cm.gray), plt.title('Diferencia en el plano R', fontsize = 15)
plt.subplot(132),plt.imshow(img[:,:,1]-imgRGB[1,:,:], cmap=plt.cm.gray), plt.title('Diferencia en el plano G', fontsize = 15)
plt.subplot(133),plt.imshow(img[:,:,2]-imgRGB[2,:,:], cmap=plt.cm.gray), plt.title('Diferencia en el plano B', fontsize = 15)
plt.show()
```

Solo fue posible implementar la conversión de OHTA a RGB puesto que no existe una conversión directa de LUX a RGB. Se nota que el resultado es aceptable, en la diferencia entre cada plano se observan unos errores minímos dados por las aproximaciones numericas de la conversión.
