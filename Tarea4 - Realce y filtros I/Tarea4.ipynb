```
import cv2
from IPython.display import Image
import numpy as np
from matplotlib import pyplot as plt
from skimage import util as sk

img = cv2.imread('/content/drive/MyDrive/ColabNotebooks/Vision_Maquina/Tarea4/giratina.jpg',0)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
N, M = img.shape[:2]               #Tamaño de la imagen
print("Tamaño de la imagen: ({}, {})".format(N,M))

plt.figure(figsize=(10,10))
plt.imshow(img, cmap=plt.cm.gray), plt.title('Imagen original', fontsize = 15)
plt.show()

noisy_img = sk.random_noise(img, mode='speckle')
plt.figure(figsize=(10,10))
plt.imshow(noisy_img, cmap=plt.cm.gray), plt.title('Imagen con ruido de tipo speckle', fontsize = 15)
plt.show()
```
