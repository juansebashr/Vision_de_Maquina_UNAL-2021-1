```python
#Calcula el CONTRASTE
B1=0; B2=0
for i in range(width):
  for j in range(height):
    B1=B1 + pow(img[i,j]-Br1,2)
    B2=B2 + pow(equ[i,j]-Br2,2);
Cont1=pow(B1/(width*height),0.5)
Cont2=pow(B2/(width*height),0.5)
#print "Contraste img. original = %d. Contraste img. ecualizada = %d." %(Cont1,Cont2)

#Visualizacion resulatado 

plt.figure(1)
plt.subplot(221), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.subplot(223), plt.hist(img.ravel(),128,[0,256])

plt.subplot(222), plt.imshow(equ, cmap='gray'), plt.title('Ecualizada')
plt.subplot(224), plt.hist(equ.ravel(),128,[0,256])

# Adjust the subplot layout, because the logit one may take more space
# than usual, due to labels 
plt.subplots_adjust(top=0.95, bottom=0.05, left=0.10, right=0.95, hspace=0.25, wspace=0.35)

plt.show()

print ("        Contraste = %d.          Contraste = %d." %(Cont1,Cont2))
```

```python
# DIVISION POR UNA CONSTANTE

# SE CARGAN LAS IMAGENES 
img1 = cv2.imread('/content/drive/MyDrive/ColabNotebooks/2_VM_Conceptos_basicos/Fig/ConceptosBasicos/Imagen_bitmap_1.jpg',0)

width, height = img1.shape[:2]
img_division = np.zeros((width,height), np.uint8)

#DIVISION POR UNA CONSTANTE
C=2 #Valor de la constante
IGMax=0
IGmin=1000000

for i in range(width):
    for j in range(height):
        img_division[i,j] = img1[i,j]/C
        if img1[i,j] > IGMax: IGMax = img1[i,j]
        if img_division[i,j] < IGmin: IGmin = img_division[i,j]

#        print (img1[i,j],img_division[i,j])
#print "Vmax= %d,   Vmin=%d." %(IGMax,IGmin)


#SE CALCULA EL HISTOGRAMA CON LA SIGUIENTE FUNCION
hist1 = cv2.calcHist([img1],[0],None,[256],[0,256])
hist_division = cv2.calcHist([img_division],[0],None,[256],[0,256])


#Visualizacion resulatado 
plt.figure(1)
plt.subplot(221), plt.imshow(img1, cmap='gray', clim=(IGmin, IGMax)), plt.title('Original')
plt.subplot(222), plt.imshow(img_division, cmap='gray', clim=(IGmin, IGMax)), plt.title('Division')

plt.subplot(223), plt.hist(img1.ravel(),256,[0,256])
plt.subplot(224), plt.hist(img_division.ravel(),256,[0,256])


# Adjust the subplot layout, because the logit one may take more space
# than usual, due to labels 
plt.subplots_adjust(top=0.95, bottom=0.05, left=0.10, right=0.95, hspace=0.25, wspace=0.35)

plt.show()

#cv2.imshow('Original',img1)
#cv2.imshow('Division',img_division)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
# OPERACIONES PUNTUALES

# SE CARGAN LAS IMAGENES 
img1 = cv2.imread('/content/drive/MyDrive/ColabNotebooks/2_VM_Conceptos_basicos/Fig/ConceptosBasicos/Imagen_bitmap_1.jpg',0)

width, height = img1.shape[:2]
img_umbral = np.zeros((width,height), np.uint8)
img_umbral_inv = np.zeros((width,height), np.uint8)
img_inv = np.zeros((width,height), np.uint8)
img_max = np.zeros((width,height), np.uint8)
img_OR = np.zeros((width,height), np.uint8)
img_AND = np.zeros((width,height), np.uint8)


#UMBRALIZACION
H=128 #Valor de la constante

for i in range(width):
    for j in range(height):
        if img1[i,j] < H : img_umbral[i,j] = 0
        else : img_umbral[i,j] = 255
        img_inv[i,j] = 255 - img1[i,j]
        img_max[i,j] = max(img_umbral[i,j], img1[i,j])
        img_OR[i,j] = img_umbral[i,j] or img_inv[i,j]
        img_AND[i,j] = img_umbral[i,j] and img1[i,j]



#Visualizacion resulatado 
plt.figure(1)
plt.subplot(331), plt.imshow(img1, cmap='gray'), plt.title('Original')
plt.subplot(332), plt.imshow(img_umbral, cmap='gray'), plt.title('Umbralizada')
plt.subplot(333), plt.imshow(img_inv, cmap='gray'), plt.title('Ori. Invertida')
plt.subplot(337), plt.imshow(img_max, cmap='gray'), plt.title('MAX(Ori,Umbr)')
plt.subplot(338), plt.imshow(img_OR, cmap='gray'), plt.title('Ori-Inv OR Umbra.')
plt.subplot(339), plt.imshow(img_AND, cmap='gray'), plt.title('Ori AND Umbra')


# Adjust the subplot layout, because the logit one may take more space
# than usual, due to labels 
plt.subplots_adjust(top=0.95, bottom=0.05, left=0.10, right=0.95, hspace=0.25, wspace=0.35)

plt.show()
```

```python
#FILTRAR RUIDO SAL Y PIMIENTA CON MEDIANA

#FILTRO MEDIANA DE TAMAO MXM
M=3   
kernel = np.zeros((M,M), np.uint8)

#CREAR IMAGEN
filas = 500
columnas = 500
img = np.zeros((filas,columnas), np.uint8)
for i in range(int(img.shape[0]/4),int(3*img.shape[0]/4)):
   for j in range(int(img.shape[1]/4),int(3*img.shape[1]/4)):
    img[i,j] = 255

ruido1 = np.zeros((filas,columnas), np.uint8)


rows=img.shape[0]-kernel.shape[0]+1
cols=img.shape[1]-kernel.shape[1]+1
output=np.zeros((rows,cols), 'uint8')


#AGREGA RUIDO SAL Y PIMIENTA
RS = 255 #valor ruido sal
RP = 0 #valor ruido pimienta
    
for i in range(filas):
    for j in range(columnas):
        VR = np.random.randint(0,100) #generar valor random
#        print "\nValor Aleatorio: %d"%(VR)
        if (VR > 90): # aplica ruido al 10% de la imagen
            if (img[i,j] == 0):
                ruido1[i,j]=RS
#                print "\n%d, Ruido sal, %d" %(img[i,j],ruido1[i,j])
            else:
                if (img[i,j] == 255):
                    ruido1[i,j]=RP
#                    print "\n%d, Ruido pimienta, %d" %(img[i,j],ruido1[i,j])
        else:
            ruido1[i,j]=img[i,j]
#            print"\n%d, NO Ruido , %d" %(img[i,j],ruido1[i,j])
            

#FILTRO MEDIANA
for i in range(0, output.shape[0]):
   for j in range(0, output.shape[1]):
        #desplazando el kernel
        Imagen_patch=img[i:i+len(kernel),j:j+len(kernel)]   
        Temp = np.array(Imagen_patch).flatten()
        Temp2 = sorted(Temp)
        output[i,j]=(int(Temp2[int(Temp.shape[0]/2)]))
    
plt.figure(1)
plt.subplot(221), plt.imshow(img, cmap='gray'), plt.title('Imagen')
plt.subplot(222), plt.imshow(ruido1, cmap='gray'), plt.title('Ruido 10%')
#plt.subplot(223), plt.imshow(ruido2, cmap='gray'), plt.title('Ruido 20%')
plt.subplot(224), plt.imshow(output, cmap='gray'), plt.title('Filtro Mediana')

# Adjust the subplot layout, because the logit one may take more space
# than usual, due to labels 
plt.subplots_adjust(top=0.95, bottom=0.05, left=0.10, right=0.95, hspace=0.25, wspace=0.35)

plt.show()
```
